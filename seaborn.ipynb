# Seaborn - statistical visualization
## what is seaborn? 
seaborn is a python library that builts on top of matplotlib that makes the statistical data more easy and beautiful.
## why we need it?
- On creating plots like violin,blox plot and pair plots.
- Automatically handles data from pandas dataframe.
## where we use it?
- Data Analysis and Reporting
- Exploring relationship b/w the variables.
- Machine learning for visualizing datasets and results.

```python
import seaborn as sns   
```

```python
# Load Titanic dataset
df = sns.load_dataset("titanic")
df.head()
```

```python
import matplotlib.pyplot as plt
# Histogram
sns.histplot(df['age'].dropna(), bins=10, kde=True)
plt.title("Histogram of Age")
plt.show()
 
# Countplot
sns.countplot(x='class', data=df)
plt.title("Count of Passengers by Class")
plt.show()
```

```python
sns.scatterplot(x='age', y='fare', data=df, hue='sex')
plt.title("Age vs Fare")
plt.show()
```

```python
### 6. Boxplot & Violin Plot
# Boxplot
sns.boxplot(x='class', y='age', data=df)
plt.title("Boxplot of Age by Class")
plt.show()
 
# Violin plot
sns.violinplot(x='class', y='age', hue='sex', data=df, split=True)
plt.title("Violin Plot of Age by Class and Sex")
plt.show()
```

```python
### 7. Pairplot
iris = sns.load_dataset('iris')
sns.pairplot(iris, hue='species')
plt.show()
```

```python
### 8. Heatmap
# Correlation heatmap (numeric columns only)
corr = iris.select_dtypes(include=['float64', 'int64']).corr()
 
import seaborn as sns
import matplotlib.pyplot as plt
 
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap of Iris Dataset")
plt.show()
```

```python
### 9. Customization & Themes
sns.set_theme(style="whitegrid")  # options: darkgrid, white, ticks, etc.
sns.histplot(df['age'].dropna(), bins=10, color='orange')
plt.title("Customized Histogram with Seaborn")
plt.show()
```

```python
sns.barplot(x='class', y='fare', data=df, hue='sex', errorbar='sd')
plt.title("Average Fare by Class and Sex")
plt.show()
```

```python
#Compare Passenger Survival by Class and Gender
df = sns.load_dataset("titanic")
sns.barplot(x='class', y='survived', data=df, hue='sex', errorbar=None)
plt.title("Passenger Survival Rate by Class and Gender")
plt.ylabel("Survival Rate")
plt.show()
```

```python
sns.histplot(df['age'].dropna(), bins=30, kde=True)
plt.title('Distribution of Passenger Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
```

```python
sns.boxplot(x='pclass', y='fare', data=df)
plt.title('Fare Distribution by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Fare ($)')
plt.show()
```

```python
sns.scatterplot(x='age', y='fare', hue='survived', data=df, alpha=0.6)
plt.title('Relationship Between Age and Fare, Colored by Survival')
plt.xlabel('Age')
plt.ylabel('Fare ($)')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

```python
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap of Titanic Numerical Features')
plt.show()
```
